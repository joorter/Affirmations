```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

def corrfunc(x, y, **kwds):
    cmap = kwds['cmap']
    norm = kwds['norm']
    ax = plt.gca()
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    sns.despine(ax=ax, bottom=True, top=True, left=True, right=True)
    r, _ = pearsonr(x, y)
    facecolor = cmap(norm(r))
    ax.set_facecolor(facecolor)
    lightness = (max(facecolor[:3]) + min(facecolor[:3]) ) / 2
    tam = int(70*abs(r))
    if tam < 10:
        tam = 10
    ax.annotate(f"{r:.2f}", xy=(.5, .5), xycoords=ax.transAxes,
                color='white' if lightness < 0.7 else 'black', 
                size=tam, ha='center', va='center')

g = sns.PairGrid(df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 
                     'AGE', 'DIS','target']])
g.map_lower(plt.scatter, s=10)
g.map_diag(sns.histplot, kde=False)
g.map_upper(corrfunc, cmap=plt.get_cmap('coolwarm'), 
            norm=plt.Normalize(vmin=-.5, vmax=.5))
g.fig.subplots_adjust(wspace=0.06, hspace=0.06) 
plt.show()
```

#### Dendrogramas

```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy import cluster
import numpy as np
np.random.seed(1234)
iris = datasets.load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X['class'] = iris.target_names[iris.target]
X = X.sample(30)
# Matriz de distancias
Z = cluster.hierarchy.ward(X[iris.feature_names])
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance (Ward)')
den = cluster.hierarchy.dendrogram(Z, labels=X['class'].values, leaf_rotation=90)
plt.show()
```

```python
# Corta por clusters
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cutree_3 = cluster.hierarchy.cut_tree(Z, n_clusters=3)
grupos = np.array(['versicolor','virginica','setosa'])
selec_cluster = grupos[cutree_3.flatten()]
reales = X['class'].values
sns.set_style('white')
cm = confusion_matrix(reales, selec_cluster)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                              display_labels=iris.target_names)
disp.plot()
plt.show()
```

```python
# Corta por altura
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cutree_3 = cluster.hierarchy.cut_tree(Z, height=4)
grupos = np.array(['versicolor','virginica','setosa'])
selec_cluster = grupos[cutree_3.flatten()]
reales = X['class'].values
sns.set_style('white')
cm = confusion_matrix(reales, selec_cluster)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                              display_labels=iris.target_names)
disp.plot()
plt.show()
```

#### Análisis de Componentes Principales (PCA)

```python
#PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import statsmodels.api as sm
USArrests = sm.datasets.get_rdataset("USArrests", "datasets")
datos = StandardScaler().fit_transform(USArrests.data)
modelo_pca = PCA().fit(datos)
componentes_principales = pd.DataFrame(modelo_pca.components_, 
       columns=USArrests.data.columns,
       index=['PC'+str(i+1) for i in range(datos.shape[1])])
display(componentes_principales)
```

```python
fig = sns.barplot(x=['PC'+str(i+1) for i in range(datos.shape[1])],
            y=modelo_pca.explained_variance_ratio_).\
set_title("Porcentaje de Varianza Explicada por cada PC")
fig = sns.lineplot(x=['PC'+str(i+1) for i in range(datos.shape[1])],
            y=modelo_pca.explained_variance_ratio_.cumsum())
print('Porcentajes de varianza explicados (acumulados):',
      modelo_pca.explained_variance_ratio_.cumsum())
```

```python
proyeccion_pca = pd.DataFrame(modelo_pca.transform(datos),
  columns=['PC'+str(i+1) for i in range(datos.shape[1])],
  index=USArrests.data.index)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))
sns.scatterplot(x='PC1',y='PC2',data=proyeccion_pca,ax=ax)
for i in range(len(proyeccion_pca)):
    ax.text(x=proyeccion_pca['PC1'].values[i], 
             y=proyeccion_pca['PC2'].values[i], 
             s=proyeccion_pca.index.values[i])
```

```python
from minisom import MiniSom
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
USArrests = sm.datasets.get_rdataset("USArrests", "datasets")
datos = StandardScaler().fit_transform(USArrests.data)
som = MiniSom(25,25,datos.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(datos)
proyeccion = np.array([[som.winner(datos[i,:])[0],
                        som.winner(datos[i,:])[1]] for i in range(len(datos))])
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))
sns.scatterplot(x=proyeccion[:,0],y=proyeccion[:,1],ax=ax)
for i in range(len(proyeccion_pca)):
    ax.text(x=proyeccion[i,0], y=proyeccion[i,1],
            s=USArrests.data.index[i])
```

#### TSNE

```python
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
from sklearn.manifold import TSNE 
USArrests = sm.datasets.get_rdataset("USArrests", "datasets")
datos = StandardScaler().fit_transform(USArrests.data)
for perp in [5,10]:
    proyeccion = TSNE(perplexity=perp).fit_transform(datos)
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))
    sns.scatterplot(x=proyeccion[:,0],y=proyeccion[:,1],ax=ax).\
        set_title(f't-SNE perplexity={perp}')
    for i in range(len(proyeccion_pca)):
        ax.text(x=proyeccion[i,0], y=proyeccion[i,1],
                s=USArrests.data.index[i])
```
